核心部分是按协议进行数据封包：
包ID：0x05
数据头：最高位=1，剩余7位对应最多7个数据的最高位
数据：数据中只保留低7为，最高位=1
校验和：暂时不用

```python
from PyQt5.QtWidgets import QApplication
from PyQt5.QtSerialPort import QSerialPort
from PyQt5.QtCore import pyqtSignal, QObject, QTimer
import struct

class SerialInterface(QObject):
    def __init__(self, port = 'COM5'):
        super().__init__()
        
        self.index = 0
        self.cmdId = 0x05

        # 串口初始化配置
        self.com = QSerialPort()
        self.com.setPortName(port) 
        self.com.setBaudRate(QSerialPort.Baud115200) 
        self.com.setDataBits(QSerialPort.Data8)
        self.com.setStopBits(QSerialPort.OneStop)
        self.com.setParity(QSerialPort.NoParity)
        self.com.setFlowControl(QSerialPort.NoFlowControl)
        
        # 打开串口
        if self.com.open(QSerialPort.ReadWrite) == False:
            print("serial port open error")
            return

        # 创建定时器，配置槽函数
        self.readtimer = QTimer()
        self.readtimer.timeout.connect(self.sendData)
        self.readtimer.start(2) # 100ms周期定时器
        
    def readData(self):
        revData = self.com.readAll()
        if revData.size() is not 0:
            print(revData)
    
    def sendData(self):
        global ecgData
        if self.index >= len(ecgData):
            self.index = 0
        
        ecg2 = ecgData[self.index]
        self.index = self.index + 1
        self.com.write(self.dataPacket(2048, ecg2, 2048, 0))
        
        self.readData()

    def dataPacket(self,ecg1,ecg2,ecg3,status):
        cmdId = 0x05
        dataHead = 0x80
        ecgTmp = bytearray(struct.pack('!3HB', ecg1, ecg2, ecg3, status))
        txData = bytearray(struct.pack('!2B', cmdId,0x80))
        for index,data in enumerate(ecgTmp):
            dataHead = dataHead | (data&0x80)>>(index+1)
            txData.append(data|0x80)
        
        txData[1] = dataHead
        txData.append(0x80) ## CRC not use
        
        return txData
```
