2020年的新冠肺炎疫情给公共医疗系统带来了极大挑战。
医疗监护技术对于判断新冠肺炎的严重程度非常重要。
以人的血氧饱和度为例，正常值在98%-100%之间，
如果新型肺炎出现重型或危重型，可能会出现低氧血症，血氧饱和度降到95%以下。
因此新闻图片上总能够看到病人床边的监护仪。

↓ 监护仪

当时发病人员极多，全国各省都派出医疗队支援湖北。
如果利用云监护技术，医护人员可同时监护多个病患，具备提效作用；
同时可有效降低医护人员进入隔离病房的次数，利于保护医护人员健康。
但是遗憾的是，此时此刻医护人员意识到，之前在远程医疗的投入仍相当缺乏。

另一方面，临时建立的方舱医院从体育馆改造，没有很好的网络条件，
远程监护仪必须支持WiFi传输，最好支持移动4G网络。
现场搭建服务器也是不现实的，现场的传染风险高，也不具备机房条件，
因此服务器应采用云计算技术，实现快速部署。

天荣公司是一家老牌的监护仪生产企业。
比国内众多的竞争对手先行一步，天荣在医院内网级别的医疗监护仪网络技术已经成熟。
对于医院这种应用环境来说，网络安全永远是第一位的。
所以之前天荣的研究集中在院内，就是怕病人的隐私数据流出医院，
迟迟未开展移动+云计算这一组合的研发工作。
但是，这场疫情来得太过突然，太过猛烈，没给人类喘息之机。
它倒逼监护仪行业推出4G+云计算的产品，而且已是箭在弦上不得不发了。

天荣为了加快开发进度，决定在原有的机型上改造，
并且委托牛艾科技开发云监护系统。
牛艾科技在物联网领域积淀深厚，专门为客户定制私有的物联网云平台。
二者一拍即合。故事就从这里开始了……

# 项目启动会 #

云监护项目被客户天荣医疗提升为2020年重点中的重点，进度要求特别急。
所以刚一复工，客户代表就来到牛艾公司，牛艾的项目经理杨大圣紧急召开启动会。
大圣把项目的背景跟在座的工程师大概叙述一遍，邀请客户代表介绍具体情况。

![](./image/cloud_monitor.png)

客户一边把PPT投影出来，一边对大家说：“这是我们天荣公司传统的医疗用监护仪的主界面。
如图，监护仪交互界面中包含：心电波形(ECG)、血氧波形(SPO2)、呼吸波形(RESP)、
心跳数据(HR)、血压数据(NIBP)、血氧数据(SPO2)、人体温度(TEMP)、呼吸数据(RESP)、二氧化碳数据(CO2)等内容。
ECG是心电图，大家应该都听说过。图上看到的红色信号SPO2血氧更常用，很多人都用过智能手环，
手环就是靠血氧信号来测量心率的。咱们看到血氧信号有周期性，一个周期就是一次心跳。
当然还有一些附件，显示设备信息、病人信息等。

单片机控制外围采样电路获得这些数据。这个部分属于模拟电路，大家不必过多关注，
因为这部分电路在我们公司每一台设备上都存在，而且电路原理也都是大同小异。
所以经过多年发展已然非常成熟，变动很微小。

前端的模拟电路把数据采集到以后，我们有一个私有的通信协议，
利用UART串口发送给上位机，这一次的项目我们希望是一台基于ARM的Linux终端。
然后在上位机解析数据，并且显示到界面上。
以上的部分跟我们传统的监护仪架构类似。”

客户翻了一张PPT，接着说：“今年的疫情大家也都清楚。
我们下到一线，与医护人员广泛沟通，眼下最大的需求是建立医生工作站。
监护仪放在病人床头，但是医生去病房的次数要减少。
因此我们要在医生办公室就能够查看数据、查看波形，一个医生能同时监管多路机器。
毕竟，你们懂得，医生太少了。你们是物联网的专家，希望各位鼎力支持，
帮助我们开发云监护仪。
云监护系统由末端监护仪获取传感器数据，并上传到云端服务器，
由服务器完成数据的存储、分析。医生可访问云端服务器，
获取病人数据，从而实现远程诊断的技术支持。”

↓ 管理工作站界面
![](./image/qt_workstation_table_demo.png)

杨大圣转过头对客户代表说：
“互联网公司分为前端和后端开发人员，
前端是指手机app或者网页的开发，后端是指数据库服务的开发，也就是服务器代码的开发。
互联网行业经常说全栈工程师难招，全栈是指前后端都能干的程序员。
但是物联网要更加复杂。物联网=互联网+硬件设备。我们常说云管端，是指：
- 云：云计算，数据库、大数据存储与处理等，运行在服务器上的。
- 管：管理界面，管理嵌入式设备，查看各类报表。
- 端：嵌入式设备，硬件部分。这种硬件是具有网络连接能力的。
- 运维：互联网需要持续提供服务，所以互联网产品往往没有开发终结的概念，
一定会不停的加入新功能，并且做许多服务器的运维工作以适应用户量的增长。”

客户不断点头，说道：“英雄所见略同。我们在这次项目需要同时实现云管端。
目前是这样规划的：
- 云：不多说了，数据库部分我们肯定需要。
- 管：在医生的工作站上安装应用软件，能够实现波形查看、设备管理等。
- 移动：也是一种管理界面，用微信小程序实现，这样医生离开医院也能够查看数据。
- 端：云监护仪，具备传统监护仪的功能以外，多了云端接口，可将本地数据上传到云端，
同时可与服务器进行必要的数据交互。”

大圣正在喝水，紧接着说：
“除了云管端，硬件设备的开发又至少分为：结构设计、硬件工程师、嵌入式应用工程师、系统移植人员。
- 结构工程师：负责外观外壳等的设计；
- 硬件工程师：负责设计电路板，有的公司又把原理图和PCB分开由不同工程师来做；
- 系统移植：移植人员负责在定制电路板上把Linux启动起来，并且开发外设驱动；
- 嵌入式工程师：负责开发Linux应用程序。”

天荣的客户代表打断说：“这些工程师岗位我们公司也有，
属于传统的嵌入式开发领域。但是我们不知道怎么开发物联网的后台，
也不知道怎么把硬件设备跟后台对接。”

杨大圣答道：“恰好是我们擅长的。给您介绍一下我们云监控工作组的工程师吧。
- 后端工程师：李凡星，负责云端即服务器代码的开发。所谓后端，主要包括物联网后台数据库，设备的历史数据存储，以及分析报告等。
- 前端工程师：肖丽，负责设备管理** 界面 **的开发，比如输入设备的配置参数啊，查看设备的数值啊，等等。管理界面根据应用场景的不同，一般会采用微信小程序、HTML5（简称H5，就是网页）。
- 嵌入式工程师：许博微，嵌入式应用程序开发，一般带屏幕的嵌入式设备采用Linux+Qt来开发。很多方面、或者说框架的本质思想上，跟PC端的C#开发也差不多。”

客户点点头，说：“听上去也不是特别复杂。”

“不，这只是一部分岗位，”大圣显得有些急躁，“属于最一线开发人员，还有很多岗位我也给您介绍一下。
- 项目经理：首先我是项目经理，负责项目进度的管理、人员的管理，以及与客户之间的直接沟通。
- 架构：付杰科，规划整个系统的架构，以及采用什么框架进行开发。
- 产品经理：黄力，负责产品功能的设定，包括用户界面等用户体验方面的设计。
产品经理的岗位很重要，因为产品在设计实现之前，其最终的型态已经在产品经理心中有了雏形。
如果产品经理的思路有问题，那么实现的再好再漂亮，客户也不会买单。
这里有一条产品哲学：问题在越早的环节被发现，成本就越低。
比如软件的bug，开发过程发现则修改成本很低。如果到客户现场才发现，那修改bug的成本就高多了。
所以Jobs、雷军这些行业领导者都自称是产品经理。
那么产品是不是越高大上越好呢？
这里往往有一个误区，不容易被认识到，因为产品越复杂开发成本肯定越高，
所以产品经理的方案必须与项目预算的实际情况相结合，不能把什么高大上的功能都往里加，
否则产品太贵，或者团队把钱花光了，也是要失败的。”

客户略有所思，“非常有道理。我们天荣公司委托你们牛艾公司开发，所以需求是我们提的。
但是你对产品经理的分析很到位。需求太复杂，里面有一些用不到的功能，只会增加成本，
关键还产生了不好的副作用，滞后开发时间，这会让我们的产品晚于竞争对手，
甚至有可能错过这个机会窗口，那样公司的损失就太大了。
所以啊，我决定先把移动软件的开发拿掉，医生们已经很累了，走出办公室就让他们休息吧。”

大圣笑了，连声答应，“那么我们剩下三个子系统需要开发：
- 云端数据库
- 管端医生工作站
- 端设备：监护仪

对吗？”

看到客户给与肯定答案，大圣说：“不过，
我们要增加Linux的移植工作，移植到对应的ARM平台。
另外，需要有白盒测试人员配合开发。”

# 方案讨论

项目组专门开了一个方案讨论会。重点从两方面开展：用什么工具实现项目需求、项目如何分工。

项目组安排四名工程师：
- 嵌入式工程师 许博微：负责从传感器获取数据并将数据转发至服务器，同时在本地把采集到的数据展示到本机的屏幕上。客户特别提出，监护仪设备有不同的版本，主要区别是供电方式和屏幕大小。这里的屏幕大小很关键，因为它涉及到我们设计出来的终端界面需要适配不同的屏幕，这在项目一开始的时候就要确认，否则开发出的软件的兼容性就有可能出问题。
  - 支持多台设备同时在线
  - Qt应用开发：绘制波形，显示多个指标
  - 发送数据；
  - 显示病人信息
- 服务器后端工程师 李凡星：负责从监护仪接收数据，并进行协议解析然后存入数据库，同时响应终端设备的数据请求，把数据库中的数据转发给终端设备。
  - 数据库
  - 数据模型设计
  - restful API实现
  - 单元测试
  - 集成测试
- 管理前端工程师 肖丽：肖丽负责H5和微信小程序开发，但是此项目第一阶段把微信小程序的功能拿掉了，只在Qt程序上实现医生工作站软件。软件功能：
  - 维护病人信息；
  - 显示设备的信息
  - 病人与设备的绑定
  - 显示监护仪数据
- 操作系统工程师：
  - 交叉编译
  - Linux内核移植
  - rootfs搭建
